```python
import os
from riotwatcher import LolWatcher, ApiError, RiotWatcher
import csv

apikey = "RGAPI-0ceaef1b-9fa2-4d30-a4a4-1d9ff2522bfb" # Inactive
region = "na1"
queue = "RANKED_SOLO_5x5"

lol_watcher = LolWatcher(api_key=apikey)

challengerQueue = lol_watcher.league.challenger_by_queue(region=region, queue=queue)["entries"][:21]
gmQueue = lol_watcher.league.grandmaster_by_queue(region=region, queue=queue)["entries"][:21]
masterQueue = lol_watcher.league.masters_by_queue(region=region, queue=queue)["entries"][:21]
```

```python
challSumNames = []
challSumId = []

gmSumNames = []
gmSumId = []

masterSumNames = []
masterSumId = []

for entry in challengerQueue:
    currSummoner = lol_watcher.summoner.by_id(region=region, encrypted_summoner_id=entry["summonerId"])

    challSumNames.append(currSummoner["name"])
    challSumId.append(currSummoner["puuid"])

for entry in gmQueue:
    currSummoner = lol_watcher.summoner.by_id(region=region, encrypted_summoner_id=entry["summonerId"])

    gmSumNames.append(currSummoner["name"])
    gmSumId.append(currSummoner["puuid"])

for entry in masterQueue:
    currSummoner = lol_watcher.summoner.by_id(region=region, encrypted_summoner_id=entry["summonerId"])

    masterSumNames.append(currSummoner["name"])
    masterSumId.append(currSummoner["puuid"])
```

```python
header = ["Game Duration", "Total Team Gold Earned", "Total Team Crowd Control Time", "Average Team Account Level", 
"Average Team Champion Mastery", "Total Team Kills", "Total Team Vision Score", "Win"]

with open('ChallengerMatchData.csv', 'w', encoding='UTF8', newline='') as f:
    writer = csv.writer(f)

    # write the header
    writer.writerow(header)

    for summoner in challSumId:
        matchlist = lol_watcher.match.matchlist_by_puuid(region="americas", puuid=summoner, type="ranked", count=25)

        for match in matchlist:
            currentMatchData = []

            currentMatch = lol_watcher.match.by_id(region="americas", match_id=match)

            currentMatchData.append(currentMatch["info"]["gameDuration"])

            team0 = currentMatch["info"]["participants"][:5]
            team1 = currentMatch["info"]["participants"][5:10]

            totalGold0 = 0
            totalGold1 = 0

            totalCC0 = 0
            totalCC1 = 0

            averageAcc0 = 0
            averageAcc1 = 0
        
            averageMas0 = 0
            averageMas1 = 0

            kills0 = 0
            kills1 = 0

            vision0 = 0
            vision1 = 0

            for participant in team0:
                totalGold0 = totalGold0 + participant["goldEarned"]
                averageMas0 = averageMas0 + participant["champExperience"]
                kills0 = kills0 + participant["kills"]
                averageAcc0 = averageAcc0 + participant["summonerLevel"]
                totalCC0 = totalCC0 + participant["timeCCingOthers"]
                vision0 = vision0 + participant["visionScore"]

            averageMas0 = averageMas0 / 5
            averageAcc0 = averageAcc0 / 5

            currentMatchData.append(totalGold0)
            currentMatchData.append(totalCC0)
            currentMatchData.append(averageAcc0)
            currentMatchData.append(averageMas0)
            currentMatchData.append(kills0)
            currentMatchData.append(vision0)    
            currentMatchData.append(int(currentMatch["info"]["teams"][0]["win"]))

            # write the data
            writer.writerow(currentMatchData)
            currentMatchData = []
            currentMatchData.append(currentMatch["info"]["gameDuration"])

            for participant in team1:
                totalGold1 = totalGold1 + participant["goldEarned"]
                averageMas1 = averageMas1 + participant["champExperience"]
                kills1 = kills1 + participant["kills"]
                averageAcc1 = averageAcc1 + participant["summonerLevel"]
                totalCC1 = totalCC1 + participant["timeCCingOthers"]
                vision1 = vision1 + participant["visionScore"]

            averageMas1 = averageMas1 / 5
            averageAcc1 = averageAcc1 / 5

            currentMatchData.append(totalGold1)
            currentMatchData.append(totalCC1)
            currentMatchData.append(averageAcc1)
            currentMatchData.append(averageMas1)
            currentMatchData.append(kills1)
            currentMatchData.append(vision1)
            currentMatchData.append(int(currentMatch["info"]["teams"][1]["win"]))

            # write the data
            writer.writerow(currentMatchData)


```
